### 1. Basic Statistics

Use Python to read in your data file and show many rows and columns your data has.

```python
import pandas as pd

# Path to the upload dataset
file_path = '/content/FAOSTAT_data_en_11-18-2024.csv'

# Loading the dataset
data = pd.read_csv(file_path)

# Displaying the number of rows and columns
rows, columns = data.shape
print(f"The dataset has {rows} rows and {columns} columns.")
```

### 2. Data Diversity

Use Python to identify how many unique values each column in your data has.  Use Python to identify which column has the greatest number of distinct values (aka the most diversity)?  If your data has a unique identifier (e.g. a person or visit or record ID) then find the column with the next most number of unique values.

```python
# Calculating the number of unique values in each column
unique_values = data.nunique()

# Find the column with the greatest number of distinct values
most_diverse_column = unique_values.idxmax()
most_distinct_values = unique_values.max()

# Display the number of unique values in each column
print("Number of unique values in each column:")
print(unique_values)

# Display the column with the greatest diversity
print(f"\nThe column with the most distinct values is '{most_diverse_column}' with {most_distinct_values} unique values.")
```

### 3. Data Redundancy

Use Python to identify any columns in your data where the value for every row is the same.  For example, if you had a gender column and all the rows said 'M', then you would want to write the code that helped you determine that.

```python
import pandas as pd
data = pd.read_csv(file_path)

# Identify columns where all values are the same
constant_columns = [col for col in data.columns if data[col].nunique() == 1]

print("Columns with the same value in all rows:")
print(constant_columns)
```

### 4. Range of Values

Your data almost certainly has some numeric columns that can be summed or averaged.  Create a histogram that shows the distribution of values for this column.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Identify numeric columns in the data
numeric_columns = data.select_dtypes(include='number').columns

# Display histograms for numeric columns
for col in numeric_columns:
    plt.figure(figsize=(10, 6))
    plt.hist(data[col].dropna(), bins=30, edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```
